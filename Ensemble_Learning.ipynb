```python
# Build models
from sklearn.model_selection import GridSearchCV
from sklearn import svm
from sklearn import tree
from sklearn.cluster import KMeans

layers = [{'models':{},'outputs':None},
          {'models':{},'outputs':None}]

# linear models
from sklearn import linear_model
layers[0]['models']['LinearRegression'] = linear_model.LinearRegression()
params = {'alpha':[0.3,0.4,0.5,0.6,0.7]}
layers[0]['models']['Ridge'] = GridSearchCV(linear_model.Ridge(), params, cv=5) 
layers[0]['models']['RidgeCV'] = linear_model.RidgeCV(alphas=[0.1, 1.0, 10.0],cv=5)
layers[0]['models']['Lasso'] = GridSearchCV(linear_model.Lasso(), params, cv=5)
layers[0]['models']['KMeans3'] = KMeans(n_clusters=3, random_state=0)
layers[0]['models']['KMeans4'] = KMeans(n_clusters=4, random_state=0)
layers[0]['models']['KMeans5'] = KMeans(n_clusters=5, random_state=0)
layers[0]['models']['KMeans6'] = KMeans(n_clusters=6, random_state=0)
layers[0]['models']['KMeans7'] = KMeans(n_clusters=7, random_state=0)

layers[1]['models']['SVR'] = svm.SVR()
layers[1]['models']['DecisionTreeRegressor'] = tree.DecisionTreeRegressor()

layers = s.fit_predict_all_layers(layers, df, y)
print(layers[1]['outputs'])

# models started producing weird warnings later in assembly line so 
# the following have been commented out until sufficient debugging

#models['OrthogonalMatchingPursuit'] = linear_model.OrthogonalMatchingPursuit()
#models['BayesianRidge'] = linear_model.BayesianRidge()

# Build out other models later, build foundation now
```
